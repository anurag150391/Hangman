```python
from random import randrange

# creating the visualization for the hangman game
hangman={0:'''
        ____________
         |''',
        1:'''
        ____________
         |
         O''',
        2:'''
        ____________
         |
         O
        /''',
        3:'''
        ____________
         |
         O
        / \\''',
        4:'''
        ____________
         |
         O
        / \\
         |''',
        5:'''
        ____________
         |
         O
        / \\
         |
        /''',
        6:'''
        ____________
         |
         O
        / \\
         |
        / \\ '''}

# dictionary of names
list_of_names = {0:"Anurag", 1:"Sanja", 2:"Aisha", 3:"Alla", 4:"Tim", 5:"Noah", 6:"Philipp", 7:"Moritz", 8:"Prabha", 9:"Katja", 10:"Katie"}

# select a random name
name_key = randrange(len(list_of_names))
name = list_of_names[name_key].upper()

# create a string of dashes to represent the hidden name
hidden_name = "-" * len(name)
print (f"Hidden name is: {hidden_name}")
guess_count = 6

# printing a divider line
print("\n","="*100, "\n")


while guess_count > 0 and hidden_name.count('-') != 0:
  index_count = 0
  picked_letter = input("Pick a letter: ").upper()
  for letter in name:
    if letter == picked_letter:
      hidden_name = hidden_name[:index_count] + picked_letter + hidden_name[index_count+1:]
    index_count += 1

  if picked_letter in name:
  # if name.count(picked_letter)>0:
    print (f"CORRECT! The name contains the letter {picked_letter}")
    print (hidden_name)
  else:
    guess_count -= 1
    print (f"WRONG! Number of guesses left: {guess_count}")
    print (hangman[6 - guess_count])

  print("\n","="*100, "\n")

if guess_count == 0:
  print ("HANGED! You lost the game!!!")
else:
  print ("CONGRATULATIONS! You have guessed the correct name!!!")
```
